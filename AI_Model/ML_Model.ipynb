```python
from flask import Flask, request
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

app = Flask(__name__)
def predict_disease(symptoms):
    #print(symptoms_dict)
    data = pd.read_csv(r"/content/drive/MyDrive/Solving For India Hackathon/Converted_Data_model.csv")
    y_train=data["disease"]
    X_train = data.iloc[:,1:]
    y_test = "HIV"

    symptoms_dict = data.iloc[0:0, 1:]
    symptoms_dict = symptoms_dict.to_dict()


    for keys in symptoms_dict:
        symptoms_dict[keys] = 0.0

    for symptom in symptoms:
        symptoms_dict[symptom] = 1.0
        print(symptoms_dict[symptom])
        print(symptoms_dict)
    
    
   
    X_test = [list(symptoms_dict.values())]
    
    #X_test = pd.DataFrame(symptoms_dict, index=[0], columns=X_train.columns)
    knn_clf = KNeighborsClassifier()
    knn_clf.fit(X_train,y_train)
    distance, prediction = knn_clf.kneighbors(X_test, n_neighbors=10)
    que = []
    for i in prediction:
        que.append(y_train.iloc[i])
   
    #prediction = knn_clf.predict(X_test)
    return que

@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        symptoms = request.form['symptoms'].split(",")
       
        prediction = predict_disease(symptoms)
        return f"Predicted disease: {prediction}"
    return '''
        <form method="post">
            Symptoms : <input type="text" name="symptoms"><br>
            <input type="submit" value="Submit"><br>
        </form>
    '''
if __name__ == '__main__':
  app.run()
```
